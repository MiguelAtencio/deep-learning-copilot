#Experiment with Gradient Descent

Implement a simple gradient descent algorithm to find the minimum value of a quadratic function $f(x) = x^2 - 4x + 4$

The goal is to understand how different learning rates affect the convergence towards the function's minimum.

Your task: Manually implement the gradient descent algorithm and observe the convergence process for at least three different learning rates.

First, you need to define the function f(x) = x^2 - 4x + 4 and its derivative.

```python
def f(x):
    return x**2 -4*x + 4

def df(x):
    return 2*x - 4
```

Add the gradient descent update rule $x = x - \eta \cdot f'(x)$, where $x$ is the current position, $\eta$ is the learning rate and $f'(x)$ is the derivative of the function at $x$.

```python
def gradient_descent(initial_x, learning_rate, epochs):
    x = initial_x
    history = []  # To store the value of x and f(x) at each step
    for _ in range(epochs):
        gradient = df(x)
        x -= learning_rate * gradient
        history.append((x, f(x)))
    return history
```

It's time to perform gradient descent for some learning rates. First, try the learning rates below. Then try higher learning rates. What happens at higher rates? Why?

```python
import matplotlib.pyplot as plt

learning_rates = [0.01, 0.1, 0.5, 1.0]
initial_x = 0  # Starting point
epochs = 10

for lr in learning_rates:
    history = gradient_descent(initial_x, lr, epochs)
    xs, ys = zip(*history)
    plt.plot(xs, label=f'LR={lr}')

plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Gradient Descent Convergence')
plt.legend()
plt.show()
```

### Solution:
Note: The minimum of the derivative should be 2

#### Learning rate of 1.0:
At this learning rate, the algorithm is bouncing back and forth the minimum because its too big.

#### Learning rate of 0.5:
At this learning rate, the algorithm was lucky enough to find the minimum at the first iteration. However, this is not reliable.

#### Learning rate of 0.1:
At this learning rate, the algorithm slowly converges to the global minimum, however, it needs more iterations

#### Learning rate of 0.01
At this learning rate, the algorithm very-slowly converges to the global minimum, resulting in even more iterations needed to converge
